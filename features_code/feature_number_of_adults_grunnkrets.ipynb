# Hypothesis

Areas with many young adults (age 27-55) will have stores with higher revenues in that area


```python
import pandas as pd
import numpy as np
```

```python
stores_train = pd.read_csv('../data/stores_train.csv')
stores_test = pd.read_csv('../data/stores_test.csv')
grunnkrets_age_distribution = pd.read_csv('../data/grunnkrets_age_distribution.csv')
```

```python

grunnkrets_age_distribution.head()
```

```python

grunnkrets_age_distribution = grunnkrets_age_distribution[grunnkrets_age_distribution.year != 2015]

grunnkrets_age_distribution.head()
```

```python
merged_file_train = stores_train.merge(grunnkrets_age_distribution, on="grunnkrets_id", how="left")
merged_file_test = stores_test.merge(grunnkrets_age_distribution, on="grunnkrets_id", how="left")
merged_file_train.shape
```

```python
merged_file_train.head()
```

```python
# TRAIN

def young_adults_in_grunnkrets():
    total_number_adults_grunnkrets = []
    counter = 0
    sum_all_grunnkrets = 0 

    for index, row in merged_file_train.iterrows():
        total_num = 0
        for i in range(27, 55):
            param = f"age_{i}"
            total_num += row[param]
        #total_number_adults_grunnkrets.append(total_num)
        if not np.isnan(total_num):
            sum_all_grunnkrets += total_num
            counter += 1
    
    average = sum_all_grunnkrets / counter


    for index, row in merged_file_train.iterrows():
        total_num = 0
        for i in range(27, 55):
            param = f"age_{i}"
            total_num += row[param]
        if np.isnan(total_num):
            total_number_adults_grunnkrets.append(average)
        else:
            total_number_adults_grunnkrets.append(total_num)
        if not np.isnan(total_num):
            sum_all_grunnkrets += total_num
            counter += 1
                
    return total_number_adults_grunnkrets

young_adults_in_grunnkrets()


instance = young_adults_in_grunnkrets()
adults_grunnkrets = pd.DataFrame()
adults_grunnkrets["number_adults_grunnkrets"] = np.asarray(instance)
adults_grunnkrets.to_csv("../features_train_csv/number_adults_grunnkrets.csv", index=False)
```
