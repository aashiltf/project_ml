```python
import pandas as pd
import numpy as np
```

```python
stores_train_df = pd.read_csv("../data/stores_train.csv")
stores_test_df = pd.read_csv("../data/stores_test.csv")
```

```python
def manipulate_df(dataframe):
    df_liste = dataframe["address"].values
    for i in range(len(df_liste)):
        j = 0
        not_treated = True 
        if not type(df_liste[i]) == float:
            while not_treated and j < len(df_liste[i]):
                if df_liste[i][j].isdigit():
                    df_liste[i] = df_liste[i][0:j]#mabye j-1
                    not_treated = False
                j += 1
    dataframe = dataframe.drop("address", axis=1)
    dataframe["address"] = df_liste
    
    return dataframe


```

```python
# TRAIN
store_train_new_address = manipulate_df(stores_train_df)
# TEST
store_test_new_address = manipulate_df(stores_test_df)
```

```python
counts = stores_train_df["address"].value_counts()
print(counts)
```

```python
mask = stores_train_df["address"].isin(counts[counts < 10].index)
stores_train_df["address"][mask] = "Other"
```

```python
dummies_train = pd.get_dummies(stores_train_df, columns=["address"], prefix="A")
dummies_test = pd.get_dummies(stores_test_df, columns=["address"], prefix="A")
```

```python
def sort_new_features(dataframe):
    dataframe_copy = dataframe.copy()
    features_unsorted = []
    for col in dataframe.columns:
        if len(col) > 1:
            if col[0:2] == "A_":
                features_unsorted.append(col)
                dataframe_copy = dataframe_copy.drop(col, axis=1)
                
    features_sorted = sorted(features_unsorted)
    for i in features_sorted:
        dataframe_copy[i] = dataframe[i].values

    return dataframe_copy
```

```python
train_sorted = sort_new_features(dummies_train)
```

```python
test_sorted = sort_new_features(dummies_test)
```
