# Hypothesis

Areas with larger revenue will have households living there with higher average income

Grunnkrets-områder med høy inntekts-gjennomsnitt, vil gjøre at samme grunnkrets-område har høyere revenue på butikker

- grunnkrets_id referer til et område
- hvis man merger grunnkrets_id i income og stores_train
- merge filer
- merge de med samme grunnkrets id


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
stores_train = pd.read_csv('../data/stores_train.csv')
stores_test = pd.read_csv('../data/stores_test.csv')
grunnkrets_income_households = pd.read_csv('../data/grunnkrets_income_households.csv')
```

```python

grunnkrets_income_households = grunnkrets_income_households[grunnkrets_income_households.year != 2015]

grunnkrets_income_households.head()

#grunnkrets_income_households.drop(labels=["year"], axis=0, ("2016")
```

```python
# Merging

merged_train = stores_train.merge(grunnkrets_income_households, on="grunnkrets_id", how="left")
merged_test = stores_test.merge(grunnkrets_income_households, on="grunnkrets_id", how="left")

```

```python
# TRAIN

def average_income_per_grunnkrets():
    avg_income_per_household = []
    parameters = ["singles", "couple_without_children", "couple_with_children", "other_households", "single_parent_with_children"]

    for index, row in merged_train.iterrows():
        mean_income = 0
        average_divide = 0
        for param in parameters:
            if (row[param] > 0):
                mean_income += row[param]
                average_divide += 1
        if (average_divide > 0):
            mean_income = mean_income / average_divide
        else: 
            mean_income = 0
        avg_income_per_household.append(mean_income)
                
    return avg_income_per_household


instance = average_income_per_grunnkrets()
avg_income = pd.DataFrame()
avg_income["average_income_per_grunnkrets"] = np.asarray(instance)
avg_income.to_csv("../features_train_csv/average_income_per_grunnkrets.csv", index=False)
```
